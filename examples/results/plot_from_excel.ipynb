```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('taxation/TAX_N400.xlsx', index_col=0)
df
```

```python
fig, ax = plt.subplots(figsize=(6,5)) #subplot_kw=dict(aspect='equal',)
_ = ax.set_xlabel(r'initial endowment $e$'); _ = ax.set_ylabel(r'disutility of working $x$')
_ = ax.set_title(r'labor supply $l$')
scatter = ax.scatter(df['e'], df['x'], c=df['l'], s=10, cmap='autumn')
_ = fig.colorbar(scatter,) #label=r'labor supply $l$')
fig.savefig('taxation/TAX_N400_labor', bbox_inches="tight", pad_inches=0.05)
```

```python
fig, ax = plt.subplots(figsize=(6,5)) #subplot_kw=dict(aspect='equal',)
_ = ax.set_xlabel(r'initial endowment $e$'); _ = ax.set_ylabel(r'disutility of working $x$')
_ = ax.set_title(r'marginal tax rate $\tau$')
scatter = ax.scatter(df['e'], df['x'], c=df['tau'], s=10, cmap='winter')
_ = fig.colorbar(scatter,) #label=r'marginal tax rate $\tau$')
fig.savefig('taxation/TAX_N400_tau', bbox_inches="tight", pad_inches=0.05)
```
