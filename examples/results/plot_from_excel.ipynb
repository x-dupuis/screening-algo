```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('TAX_N400_2023-09-28_00-44-17.xlsx', index_col=0)
df
```

```python
fig, ax = plt.subplots(figsize=(5,5)) #subplot_kw=dict(aspect='equal',)
_ = ax.set_xlabel(r'initial endowment $e$'); _ = ax.set_ylabel(r'disutility of working $x$')
# _ = ax.set_title(title)
scatter = ax.scatter(df['e'], df['x'], c=df['l'], s=10, cmap='autumn')
_ = fig.colorbar(scatter, label='labor supply $l$')
# fig.savefig('TAX_N400_labor', bbox_inches="tight", pad_inches=0.05)
```
